### Q1. Grid Builder (10 pts)
Write a function `make_grid(rows, cols)` that returns a **2D list** with the given size,
filled with the string `"."`.

Example:
- `make_grid(2, 4)` returns:  
[  
[".", ".", ".", "."],  
[".", ".", ".", "."]  
]

```python
# Q1 answer here
def make_grid(rows: int, cols: int) -> list[list[str]]:
  rows, cols = 2, 4
  grid = []
  for i in range(rows):
    row = []
    for c in range(cols):
      row.append(".")
    grid.append(row)
  return grid

# TODO

# quick check
print(make_grid(2, 4))
```


### Q2. Pretty Printer (15 pts)
Write a function `print_grid(grid)` that prints each row on its own line.

Rules:
- Use **nested loops**
- Print items in a row separated by a space

Example output for a 2×3 grid of dots:
. . .
. . .

```python
# Q2 here

def print_grid(grid: list[list[str]]) -> None:
    # TODO (nested loops required)
    for i in range(len(grid)):
      for j in range(len(grid[i])):
        print(grid[i][j], end=" ")
      print()

# quick check
g = make_grid(2, 3)
print_grid(g)
```

### Q3. Count Values in a 2D List (15 pts)
Write a function `count_value(grid, target)` that returns how many times `target`
appears in the 2D list.

Rules:
- Use **nested loops**
- No dictionaries

Example:
If grid is:  

X . .  
. X .  
. . X  

Then `count_value(grid, "X")` returns `3`.

```python
def count_value(grid: list[list[str]], target: str) -> int:
    # TODO (nested loops required)
    #pass
    count = 0
    for i in range(len(grid)):
      for j in range(len(grid[i])):
        if grid[i][j] == target:
          count+=1
    return count
print(count_value([[".","x","c"],["y",".","."]], "."))
```

### Q4. Seating Chart Mini-Task (20 pts)
You are given a 4×5 seating chart filled with `"EMPTY"`.
Write the following functions:

1. `assign_seat(chart, row, col, name)`  
   Places `name` at `chart[row][col]` **only if** the seat is `"EMPTY"`.  
   Returns `True` if successful, `False` otherwise.

2. `empty_count(chart)`  
   Returns the number of seats still equal to `"EMPTY"`.

Rules:
- Use **nested loops** in `empty_count`
- No dictionaries, no lambdas, no regex


```python
# Q4 here
def assign_seat(chart: list[list[str]], row: int, col: int, name: str) -> bool:
    # TODO
    for row in range(len(chart)):
      for col in range(len(chart[row])):
        if chart[row][col] == "EMPTY":
          chart[row][col] = name
          return True
        else:
          return False

def empty_count(chart: list[list[str]]) -> int:
    # TODO (nested loops required)
    count = 0
    for row in range(len(chart)):
      for col in range(len(chart[row])):
        if chart[row][col] == "EMPTY":
          count += 1
    return count

# quick check
chart = [["EMPTY"] * 5 for _ in range(4)]
print(assign_seat(chart, 0, 0, "Ava"))
print(assign_seat(chart, 0, 0, "Mia"))  # should be False
print(empty_count(chart))
```

### Q5. Multiplication Table (15 pts)
Write a function `mult_table(n)` that prints an `n × n` multiplication table.

Rules:
- Use **nested loops**
- `n` will be at least 1
- Print one row per line

Example for `n = 4`:  
1 2 3 4  
2 4 6 8  
3 6 9 12  
4 8 12 16  

```python
# Q5 here
def mult_table(n: int) -> None:
    # TODO (nested loops required)
    grid = []
    num = 1
    for i in range(n):
      row = []
      for j in range(n):
        row.append(num)
        num +=1
      grid.append(row)

    for i in range(n):
      for j in range(n):
        print(grid[i][j], end =' ')
      print()


# quick check
mult_table(4)
```


### Q6. Find the Largest Number in a 2D List (15 pts)
Write a function `max_in_grid(nums)` that returns the largest integer in a 2D list.

Rules:
- Use **nested loops**
- Assume the grid has at least one number

Example:
If `nums = [[1, 9], [3, 4]]` then result is `9`.

```python
# Q6 here
def max_in_grid(nums: list[list[int]]) -> int:
    # TODO (nested loops required)
    largest = 0
    for i in range(len(nums)):
      for j in range(len(nums[i])):
        if nums[i][j] > largest:
          largest = nums[i][j]
    return largest

# quick check
print(max_in_grid([[1, 9], [3, 4]]))
```

### Q7. Pattern Printer (10 pts)
Write a function `staircase(n)` that prints a staircase made of `#`.

Rules:
- Use nested loops
- If `n = 5`, print:


```python
# Q7 here
def staircase(n: int) -> None:
    # TODO (nested loops required)
    grid = []
    num = 1
    for i in range(n):
      row = []
      for j in range(num):
        row.append("#")
      num +=1
      grid.append(row)

# quick check
print(staircase(5))
```
