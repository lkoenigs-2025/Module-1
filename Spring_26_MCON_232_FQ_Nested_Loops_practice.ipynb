## Nested Lists (2D Lists) — Quick Scaffolding

A **nested list** (also called a **2D list**) is a list where each item is itself a list.
We often use nested lists to represent **grids** like seating charts, game boards, or tables.

### 1) A small example (2 rows × 3 columns)

```python
grid = [
    ["EMPTY", "EMPTY", "EMPTY"],
    ["EMPTY", "EMPTY", "EMPTY"]
]
```
### 2) Building a grid with loops (recommended)

To build a grid, we usually make rows one at a time and add them to the chart.
```
def make_chart(rows, cols):
    chart = []
    for r in range(rows):              # outer loop = rows
        row = []
        for c in range(cols):          # inner loop = columns
            row.append("EMPTY")
        chart.append(row)
    return chart
```

### 3) Printing a grid neatly (nested loops)

```
def print_chart(chart):
    for r in range(len(chart)):
        for c in range(len(chart[0])):
            print(chart[r][c], end=" ")
        print()  # new line after each row
```

### 4) Updating one cell
```
chart[2][3] = "Ava"   # row 2, col 3
```

### 5) Common reminders

- The outer loop usually goes through the rows

- The inner loop usually goes through the columns

- Use len(chart) for number of rows

- Use len(chart[0]) for number of columns (assuming all rows are same length)

### F1. Seating Chart (nested list) (2D list + loops + functions)
Create a seating chart using a nested list (list of rows).

**Requirements:**
- Build a 4×5 chart filled with the string `"EMPTY"`
- Write a function `assign_seat(chart, row, col, name)` that places a name
- Write a function `print_chart(chart)` that prints each row neatly
- Assign at least 6 names to different seats
- Count how many seats are still `"EMPTY"` and print that number

**Optional extension:** Write the seating chart to a file (one row per line).


````python
# F1 code here
chart = []
row,col = 4,5
name = ["Leah", "Baila", "Chaya", "Atara", "Sarah", "Esti"]

#create a chart - nested lists
for r in range(row):
  row = []
  for c in range(col):
    #append to the row
    row.append("EMPTY")
  #append the row to the chart
  chart.append(row)

#function that assigns a name to some of the seats
def assign_seat(chart, row, col, name):
  num = 0
  #loop through each value in the nested list
  for i in range(len(chart)):
    #add a name only if num is less than 6
    if num < 6:
      chart[i][0] = name[i]
      num += 1
    if num < 6:
      chart[i][1] = name[i+3]
      num += 1

#function that prints chart and number of seats empty
def print_chart(chart):
  #accumulator variable
  count = 0
  #loop through nested lists
  for i in range(len(chart)):
    for j in range(len(chart[i])):
      #print row on each line
      print(chart[i][j], end=' ')
      #count number of seats still empty
      if chart[i][j] == "EMPTY":
        count+=1
    print()
  print("Number of seats empty:", count)

#call functions
assign_seat(chart, row, col, name)
print_chart(chart)
````

### F2. Temperature Grid (nested list) (2D list + loops + functions)
Create a weekly temperature tracker as a nested list.

**Requirements:**
- Build a 7×3 grid (7 days, 3 readings per day: morning/afternoon/evening) filled with `0`
- Write a function `set_temp(grid, day, time_index, temp)` that updates one value  
  (`time_index`: 0 = morning, 1 = afternoon, 2 = evening)
- Write a function `print_grid(grid)` that prints each day neatly
- Fill in at least **10** temperatures (your choice)
- Use nested loops to compute:
  - the **overall average temperature**
  - the **highest single reading** in the grid

**Optional extension:** Count how many readings are above a threshold (e.g., above 75).


````python
# F2 answer here
day, time_index = 7,3
grid = []
temp = 0

#create a nested list
for i in range(day):
  day = []
  for t in range(time_index):
    #append a 0 to the day list
    day.append(0)
  #append row to whole grid
  grid.append(day)

#function that
def set_temp(grid, day, time_index, temp):
  highest = -1
  total = 0
  num = 0
  #loop through the nesed list
  for d in range(len(grid)):
    for t in range(len(day)):
      #fill in at least 10 temperatures
      if num < 10:
        temp += 12
        #change the value
        grid[d][t] = temp
        num +=1
      #get the highest temp
      if temp > highest:
        highest = temp

      #total of all temps
      total += grid[d][t]
    average = total / len(grid)
  print("Highest:", highest)
  print(f'Average Temperature: {average:.2f}')

#function that prints grid
def print_grid(grid):
  for i in range(len(grid)):
    print("Day #", i+1, ":", grid[i][0], ",", grid[i][1], ",", grid[i][2])

#call functions
set_temp(grid, day, time_index, temp)
print_grid(grid)
````

### F3. Mini Gradebook (nested list) (2D list + loops + functions)
Create a mini gradebook for students and assignments using a 2D list.

**Requirements:**
- Build a 5×4 grid (5 students, 4 assignments) filled with `0`
- Write a function `set_score(gradebook, student, assignment, score)` that updates a score
- Write a function `print_gradebook(gradebook)` that prints each student’s scores neatly
- randomize grades selecting 4 in a range of 65-100  
- Use nested loops to compute and return tuple:
  - each student’s **average score**
  - the **overall class average**
  - Find the student (row) with the highest total score.
  - print returned tuple using an f-string




````python
# F3 answer here
import random

#create a nested list
gradebook = []
student, assignment = 5, 4
for s in range(student):
  score = []
  for a in range(assignment):
    #append to the row
    score.append(0)
  #append to the grid
  gradebook.append(score)

#function that updates score
def set_score(gradebook, student, assignment, score):
  total = 0
  clss = 0
  highest = -1
  for g in range(len(gradebook)):
    for i in range(len(gradebook[g])):
      gradebook[g][i] = random.randint(65, 100)
      total += gradebook[g][i]
    #find highest student score
    if total > highest:
      highest = gradebook[g][i]
    #average score
    average = total / student
    print(f'Average of Student {g+1}: {average:,.0f}')
    clss += total
    total = 0
  clss_avg = clss / (student*assignment)
  print()
  print(f'Class Average: {clss_avg:,.0f}, Highest total score: {highest} ')
  print()

def print_gradebook(gradebook):
  for g in range(len(gradebook)):
    print("Student",g+1," Grades:", end=' ')
    for i in range(len(gradebook[g])):
      print(gradebook[g][i], end='  ')
    print()

set_score(gradebook, student, assignment, score)
print_gradebook(gradebook)
````

- Use nested loops to count how many of your drawing symbols appear in the grid

**Optional extension:** Draw a border around the grid using `"#"`.


````python
# F4 answer here
grid = []
row, col = 8, 8
symbl = tuple("#$%@^!&*()")
for r in range(row):
  symbol = []
  for c in range(col):
    symbol.append(".")
  grid.append(symbol)
def draw_pixel(grid, row, col, symbol):
  num = 0
  count = 0
  for i in range(len(grid)):
    for r in range(len(symbol)):
      if num >= 12:
        grid[i][r] = "."
      else:
        grid[i][r] = random.choice(symbl)
        num +=1
        count+=1
  print("Number of symbols:", count)
  print()
def print_art(grid):
  for g in range(len(grid)):
    for s in range(len(symbol)):
      print(grid[g][s], end='  ')
    print()
draw_pixel(grid, row, col, symbol)
print_art(grid)
````
