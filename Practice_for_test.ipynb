```python
def make_board(rows,cols):
  board = []
  for r in range(rows):
    row = []
    for  c in range(cols):
      row.append("O")
    board.append(row)
  return board

print(make_board(2,3))
```

```python
def print_board(board):
  for r in range(len(board)):
    print("|", end=" ")
    for c in range(len(board[r])):
      print(board[r][c], end= " ")
    print("|")
    print()

print_board(make_board(2,3))
```

```python
board = [["O", "X", "O"], ["X", "O", "X"], ["O", "O", "X"]]
def count_symbol(board, symbol):
  count = 0
  for i in range(len(board)):
    for j in range(len(board[i])):
      if symbol == board[i][j]:
        count+=1
  return count

print(count_symbol(board,"O"))
```

```python
board = [["O", "X", "O"], ["X", "O", "X"]]
def replace_value(board, old, new):
  for i in range(len(board)):
    for j in range(len(board[i])):
      if board[i][j] == old:
        board[i][j] = new
  return board
print(replace_value(board, "X", "O"))
```

```python
def triangle(n):
  grid = []
  for i in range(n+1):
    for j in range(i):
      row = ["*"] *i
      grid.append(row)
      break

  for i in range(len(grid)):
    for j in range(len(grid[i])):
      print(grid[i][j], end=" ")
    print()

triangle(5)
```

```python
grid = [["X", ".", "."], [".", ".", "."], [".", "X", "."]]
def rows_with_x(grid):
  count = 0
  for i in range(len(grid)):
    for j in range(len(grid[i])):
      if grid[i][j] == "X":
        count+=1
        break
  return count

rows_with_x(grid)
```

```python
chart = [["A", "B", "EMPTY"], ["C", "D", "EMPTY"]]
def assign_next(chart, name):
  for i in range(len(chart)):
    for j in range(len(chart[i])):
      if chart[i][j] == "EMPTY":
        chart[i][j] = name
        return True
  return False

assign_next(chart, "John")
```

```python
def checkboard(n):
  grid = []
  for i in range(n):
    row = []
    for j in range(n):
      if (i+j) % 2 == 0:
        row.append("X")
      else:
        row.append("O")
    grid.append(row)
  return grid

print(checkboard(5))
```

```python
def number_triangle(n):
  for i in range(1, n+1):
    for j in range(1, i+1):
      print(j, end=" ")
    print()

number_triangle(5)
```

```python
def reverse_triangle(n):
  for i in range(n, 0, -1):
    for j in range(i):
      print("*", end=" ")
    print()
reverse_triangle(5)
```

```python
def fill_grid(rows, cols):
  grid = []
  for i in range(rows):
    row = []
    for j in range(cols):
      row.append(".")
    grid.append(row)
  return grid
fill_grid(2,4)
```

```python
grid = [['.', '.', '.'], ['.', 'X', '.'], ['.', '.', '.']]
def find_symbol(grid, symbol):
  for i in range(len(grid)):
    for j in range(len(grid[i])):
      if grid[i][j] == symbol:
        return (1,1)
  return (-1,-1)

find_symbol(grid, "X")
```

```python
def border(n):
  for i in range(n):
    for j in range(n):
      if i == 0 or i == 4:
        print("#", end = " ")
      elif j == 0 or j == 4:
        print("#", end = " ")
      else:
        print(".", end =" ")
    print()

border(5)
```
