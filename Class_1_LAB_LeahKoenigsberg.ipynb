```python
#create a list
nums=[1,3,5]

#1. SIMPLE FOR LOOP:
print()
print('1) SIMPLE FOR LOOP:')
for i in nums:
  print(i)

#2. LEN FOR LOOP:
print()
print('2) LEN FOR LOOP:')
for i in range(len(nums)):
  print(nums[i])

#3. ENUMERATE FOR LOOP:
print()
print('3) ENUMERATE FOR LOOP:')
for i, value in enumerate(nums):
  print(f'{i} {nums[i]} {value}')

#4. WHILE LOOP- ACCUMULATOR:
print()
print('4) WHILE LOOP- ACCUMULATOR:')
index = 0
while index < len(nums):
  print(nums[index])
  index+=1
```

```python
#Multidemesional lists- matrix
rows=4
cols=3
multi = []
val = 1

#can use an undescore
for _ in range(rows):
  row = []
  for _ in range(cols):
    row.append(val)
  multi.append(row)
  val+=1


print(multi)
```
